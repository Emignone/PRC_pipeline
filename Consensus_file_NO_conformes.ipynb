## ECR

```python
def calc_ecrs(dir_excel_5p):
    
    excel_5p = dir_excel_5p
    
    exp_dock = excel_5p['exp_dock6'].tolist()
    exp_icm = excel_5p['exp_icm'].tolist()
    exp_ad = excel_5p['exp_ad'].tolist()
    exp_pl = excel_5p['exp_pl'].tolist()
    exp_rd = excel_5p['exp_rd'].tolist()
    exp_vina = excel_5p['exp_vina'].tolist()
    
    #calc ecr 2p
    excel_5p ["ecr_ad_dock6"] = calc_ecr2(exp_ad,exp_dock,dir_excel_5p)
    excel_5p ["ecr_ad_rd"] = calc_ecr2(exp_ad,exp_rd,dir_excel_5p)
    excel_5p ["ecr_ad_icm"] = calc_ecr2(exp_ad,exp_icm,dir_excel_5p)
    excel_5p ["ecr_ad_pl"] = calc_ecr2(exp_ad,exp_pl,dir_excel_5p)
    excel_5p ["ecr_dock6_rd"] = calc_ecr2(exp_dock,exp_rd,dir_excel_5p)
    excel_5p ["ecr_dock6_icm"] = calc_ecr2(exp_dock,exp_icm,dir_excel_5p)
    excel_5p ["ecr_dock6_pl"] = calc_ecr2(exp_dock,exp_pl,dir_excel_5p)
    excel_5p ["ecr_rd_icm"] = calc_ecr2(exp_rd,exp_icm,dir_excel_5p)
    excel_5p ["ecr_rd_pl"] = calc_ecr2(exp_rd,exp_pl,dir_excel_5p)
    excel_5p ["ecr_icm_pl"] = calc_ecr2(exp_icm,exp_pl,dir_excel_5p)
    excel_5p["ecr_ad_vina"] = calc_ecr2(exp_ad, exp_vina, dir_excel_5p)
    excel_5p["ecr_dock6_vina"] = calc_ecr2(exp_dock, exp_vina, dir_excel_5p)
    excel_5p["ecr_rd_vina"] = calc_ecr2(exp_rd, exp_vina, dir_excel_5p)
    excel_5p["ecr_icm_vina"] = calc_ecr2(exp_icm, exp_vina, dir_excel_5p)
    excel_5p["ecr_pl_vina"] = calc_ecr2(exp_pl, exp_vina, dir_excel_5p)
    
    #calc ecr 3p
    excel_5p ["ecr_ad_dock6_rd"] = calc_ecr3(exp_ad,exp_dock,exp_rd,dir_excel_5p)
    excel_5p ["ecr_ad_dock6_icm"] = calc_ecr3(exp_ad,exp_dock,exp_icm,dir_excel_5p)
    excel_5p ["ecr_ad_dock6_pl"] = calc_ecr3(exp_ad,exp_dock,exp_pl,dir_excel_5p)
    excel_5p ["ecr_dock6_rd_icm"] = calc_ecr3(exp_dock,exp_rd,exp_icm,dir_excel_5p)
    excel_5p ["ecr_dock6_rd_pl"] = calc_ecr3(exp_dock,exp_rd,exp_pl,dir_excel_5p)
    excel_5p ["ecr_rd_icm_pl"] = calc_ecr3(exp_rd,exp_icm,exp_pl,dir_excel_5p)
    excel_5p ["ecr_ad_rd_icm"] = calc_ecr3(exp_ad,exp_rd,exp_icm,dir_excel_5p)
    excel_5p ["ecr_ad_rd_pl"] = calc_ecr3(exp_ad,exp_rd,exp_pl,dir_excel_5p)
    excel_5p ["ecr_dock6_icm_pl"] = calc_ecr3(exp_dock,exp_icm,exp_pl,dir_excel_5p)
    excel_5p ["ecr_ad_icm_pl"] = calc_ecr3(exp_ad,exp_icm,exp_pl,dir_excel_5p)

    excel_5p['ecr_pl_rd_vina'] = calc_ecr3(exp_pl,exp_rd,exp_vina,dir_excel_5p)
    excel_5p['ecr_ad_dock6_vina'] = calc_ecr3(exp_ad,exp_dock,exp_vina,dir_excel_5p)
    excel_5p['ecr_ad_rd_vina'] = calc_ecr3(exp_ad,exp_rd,exp_vina,dir_excel_5p)
    excel_5p['ecr_ad_pl_vina'] = calc_ecr3(exp_ad,exp_pl,exp_vina,dir_excel_5p)
    excel_5p['ecr_dock6_rd_vina'] = calc_ecr3(exp_dock,exp_rd,exp_vina,dir_excel_5p)
    excel_5p['ecr_dock6_pl_vina'] = calc_ecr3(exp_dock,exp_pl,exp_vina,dir_excel_5p)


    #calc ecr 4p
    excel_5p ["ecr_ad_dock6_rd_icm"] = calc_ecr4(exp_ad,exp_dock,exp_rd,exp_icm,dir_excel_5p)
    excel_5p ["ecr_ad_dock6_rd_pl"] = calc_ecr4(exp_ad,exp_dock,exp_rd,exp_pl,dir_excel_5p)
    excel_5p ["ecr_ad_dock6_icm_pl"] = calc_ecr4(exp_ad,exp_dock,exp_icm,exp_pl,dir_excel_5p)
    excel_5p ["ecr_ad_rd_icm_pl"] = calc_ecr4(exp_ad,exp_rd,exp_icm,exp_pl,dir_excel_5p)
    excel_5p ["ecr_dock6_rd_icm_pl"] = calc_ecr4(exp_dock,exp_rd,exp_icm,exp_pl,dir_excel_5p)

    excel_5p["ecr_ad_dock6_rd_vina"] = calc_ecr4(exp_ad, exp_dock, exp_rd, exp_vina, dir_excel_5p)
    excel_5p["ecr_ad_dock6_pl_vina"] = calc_ecr4(exp_ad, exp_dock, exp_pl, exp_vina, dir_excel_5p)
    excel_5p["ecr_ad_pl_rd_vina"] = calc_ecr4(exp_ad, exp_rd, exp_pl, exp_vina, dir_excel_5p)
    excel_5p["ecr_dock6_pl_rd_vina"] = calc_ecr4(exp_dock, exp_rd, exp_pl, exp_vina, dir_excel_5p)

    
    #calc ecr 5p
    #con icm
    excel_5p ["ecr_ad_dock6_rd_icm_pl"] = calc_ecr5(exp_ad,exp_dock,exp_rd,exp_icm,exp_pl,dir_excel_5p)
    #con vina
    excel_5p ["ecr_ad_dock6_pl_rd_vina"] = calc_ecr5(exp_ad,exp_dock,exp_rd,exp_vina,exp_pl,dir_excel_5p)


    
    
    excel_5p.to_excel(target_dir+target+'_consensus_5p.xlsx')
```

```python
def calc_ecr2(exp_p1,exp_p2,dir_excel_5p):

    excel_temp = (dir_excel_5p)
    
    zipped_lists_2 = zip(exp_p1, exp_p2)
    ecr_2 = [x1 + x2  for (x1, x2) in zipped_lists_2] # para q sume los de la misma linea
    excel_temp['exp_p1_p2'] = ecr_2
    
    # Agrego la columna del rank exp y vuelvo a ordenar por mol
    df_temp = excel_temp.sort_values(by =['exp_p1_p2'], axis=0, ascending=False)
    cons_rank = np.arange(1, len(df_temp)+1)
    df_temp['ECR_2p']=cons_rank
    df_temp = df_temp.sort_values(by=['IX'], axis=0)
    df_temp.drop(columns=['exp_p1_p2'])
    
    return pd.to_numeric(df_temp['ECR_2p'].tolist())
```

```python
def calc_ecr3(exp_p1,exp_p2,exp_p3,dir_excel_5p):

    excel_temp = (dir_excel_5p)
    
    zipped_lists_3 = zip(exp_p1, exp_p2, exp_p3)
    ecr_3 = [x1 + x2 + x3 for (x1, x2, x3) in zipped_lists_3] # para q sume los de la misma linea
    excel_temp['exp_p1_p2_p3'] = ecr_3
    
    # Agrego la columna del rank exp y vuelvo a ordenar por mol
    df_temp = excel_temp.sort_values(by =['exp_p1_p2_p3'], axis=0, ascending=False)
    cons_rank = np.arange(1, len(df_temp)+1)
    df_temp['ECR_3p']=cons_rank
    df_temp = df_temp.sort_values(by=['IX'], axis=0)
    df_temp.drop(columns=['exp_p1_p2_p3'])
    
    return pd.to_numeric(df_temp['ECR_3p'].tolist())
```

```python
def calc_ecr4(exp_p1,exp_p2,exp_p3,exp_p4,dir_excel_5p):

    excel_temp = (dir_excel_5p)
    
    zipped_lists_4 = zip(exp_p1, exp_p2, exp_p3, exp_p4)
    ecr_4 = [x1 + x2 + x3 + x4 for (x1, x2, x3, x4) in zipped_lists_4] # para q sume los de la misma linea
    excel_temp['exp_p1_p2_p3_p4'] = ecr_4
    
    # Agrego la columna del rank exp y vuelvo a ordenar por mol
    df_temp = excel_temp.sort_values(by =['exp_p1_p2_p3_p4'], axis=0, ascending=False)
    cons_rank = np.arange(1, len(df_temp)+1)
    df_temp['ECR_4p']=cons_rank
    df_temp = df_temp.sort_values(by=['IX'], axis=0)
    df_temp.drop(columns=['exp_p1_p2_p3_p4'])
    
    return pd.to_numeric(df_temp['ECR_4p'].tolist())
```

```python
def calc_ecr5(exp_p1,exp_p2,exp_p3,exp_p4,exp_p5,dir_excel_5p):

    excel_temp = (dir_excel_5p)
    
    zipped_lists_5 = zip(exp_p1, exp_p2, exp_p3, exp_p4, exp_p5)
    ecr_5 = [x1 + x2 + x3 + x4 + x5 for (x1, x2, x3, x4, x5) in zipped_lists_5] # para q sume los de la misma linea
    excel_temp['exp_p1_p2_p3_p4_p5'] = ecr_5
    
    # Agrego la columna del rank exp y vuelvo a ordenar por mol
    df_temp = excel_temp.sort_values(by =['exp_p1_p2_p3_p4_p5'], axis=0, ascending=False)
    cons_rank = np.arange(1, len(df_temp)+1)
    df_temp['ECR_5p']=cons_rank
    df_temp = df_temp.sort_values(by=['IX'], axis=0)
    df_temp.drop(columns=['exp_p1_p2_p3_p4_p5'])
    
    
    return pd.to_numeric(df_temp['ECR_5p'].tolist())
```
